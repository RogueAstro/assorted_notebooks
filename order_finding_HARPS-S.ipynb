```python
import numpy as np
import ccdproc as cp
import matplotlib.pyplot as plt
import astropy.units as u
import scipy.signal as ss

%matplotlib notebook
```

```python
ccd = cp.fits_ccddata_reader('HARPS.2009-05-12T00_55_28.664.fits', unit=u.adu)
```

```python
plt.plot(ccd.data[2048])
plt.xlim(0, 2048)
plt.ylabel('Flux (ADU)')
plt.xlabel('Pixel')
plt.show()
```

```python
peakind = ss.find_peaks_cwt(ccd.data[2048], widths=np.arange(5,15))

plt.plot(ccd.data[2048])

for pk in peakind:
    plt.plot(pk, ccd.data[2048, pk], '*', color='r')
plt.xlim(0, 2048)
plt.ylabel('Flux (ADU)')
plt.xlabel('Pixel')
plt.show()
```
